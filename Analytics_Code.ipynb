## Step 7 Proportion Analytics

```python
import numpy as np
import pandas as pd
```

```python
# read data or using Epoch_df
#out = pd.read_csv('/Users/chenyuli/Research_Thesis/Analytics/CI_Calculation.csv')
out = Epoch_out
out
```

```python
#Change value 0 to 0.5 / Arc sin transform
out_transform=out.replace(0,0.5)
out_transform
```

```python
interval = []
estimation = []
err_range = []
for i in range(1, 4):
    interval.append([])
    estimation.append([])
    err_range.append([])
    for j in range(1, 6):
        pHat = out_transform.iloc[i, j] / out_transform.iloc[0, j]
        lower = pHat - 1.96 * np.sqrt(pHat * (1 - pHat)) / np.sqrt(out_transform.iloc[0, j])
        if (lower < 0):
            lower=0
        higher = pHat + 1.96 * np.sqrt(pHat * (1 - pHat)) / np.sqrt(out_transform.iloc[0, j])
        err=1.96 * np.sqrt(pHat * (1 - pHat)) / np.sqrt(out_transform.iloc[0, j])
        
        interval[i - 1].append((round(lower, 2), round(higher, 2)))
        estimation[i - 1].append(round(pHat, 2))
        err_range[i - 1].append(round(err,3))
```

```python
interval = pd.DataFrame(interval)
interval.columns =('2010-2013', '2014-2016', '2017', '2018', '2019')
interval.index= ('Real-World_Method', 'Sensitivity_Analysis','Handled_Missing_Data')
```
